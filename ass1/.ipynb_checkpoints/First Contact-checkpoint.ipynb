```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
test_target = pd.read_csv('data/test_target.csv')
```

```python
def elevate_factors(x,n):
    y = x
    for i in range(2,n+1):
        x_elevated = np.power(x,i)
        y = np.concatenate((y,x_elevated),axis=1)
    return y
```

```python
x_train = train.drop(['shares','url'], axis=1).values
x_test = test.drop(['url'], axis=1).values
y_test = test_target.values
y_train = train['shares']
```

```python
x_train = elevate_factors(x_train,2)
x_train = np.concatenate((np.ones((x_train.shape[0],1)),x_train),axis=1)
x_train.shape
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_train
```

```python
teta = 10*np.random.rand(x_train.shape[1])
m = x_train.shape[0]
a = 0.2

for i in range(10000):
    yhat = np.dot(teta,x_train.transpose())
    J = (1/(2*m))*np.sum(np.power((yhat - y_train), 2))
    print('Iteration {}'.format(i))
    print('Cost: %e' % (J))
    dteta = (1/m)*(np.dot((yhat - y_train),x_train))
    teta -= a*dteta
    print('--------------------')
```
